Plotting the posterior distributions: the analytical function, and Gaussian approximations with λ and logλ

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l = np.linspace(0.0001,10,100) 
r = 2
p0 = np.exp(-l)*(l**(r-1))/np.math.factorial(r) #r or r-1? 
p1 = 1 /(np.sqrt(2*np.pi*(r-1))) * np.exp(-((l-(r-1))**2)/(2*(r-1)))
p2 = (np.sqrt((r)/(2*np.pi)))*(1/l)*np.exp(-(r/2)*((np.log(l)-np.log((r)))**2))
plt.plot(l,p0/(np.sum(p0)))
plt.plot(l,p1/(np.sum(p1)))
plt.plot(l,p2/(np.sum(p2)))
plt.legend(['p0','p1','p2'])
plt.xlabel('λ')
plt.ylabel('p(λ)')
plt.show()
```
